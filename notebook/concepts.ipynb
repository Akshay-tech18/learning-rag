## Tokenization

Computers cannot read plain text, they process numbers. 
Tokenization is the first step in NLP where the process of chopping down a string into the smaller, manageable pieces called tokens.                       
Action: "The color of the sky is" gets sliced into distinct tokens.             
Result: ["The", "color", "of", "the", "sky", "is"] (Note: Each of these tokens is actually assigned an integer ID from the model's vocabulary list).

## Embeddings

Once the text is tokenized, we need to translate those tokent into the numbers.
Embeddings are dense arrays(vectors) of the real numbers that capture the semantic meaning of the token.                                                    
The word "Apple" may be represented as the [0.85,-0.22, 0.91, ... etc].
Model learns relationships mathematically.                                      
Because Transformers process everything simultaneously, they inherently do not know what order the words are in. To fix this, the model injects a Positional Encoding into each embedding. This adds a slight mathematical tweak to the vector to say, "I am the word 'sky', and I am the 5th word in this sequence."


## Semantic Similarity

Semantic similarity measures how closely related two pieces of text are in meaning, regardless of whether they use the exact same words.                    
By converting two sentences into their mathematical embeddings we can measure the distance or angle between them in our high dimensional space. we use the metric cosine similarity here. A high similarity score means the sentences means roughly the same thing

## context window

The context window is the models working memory. It is the maximum number of tokens a model can consider at one single time when reading a prompt and generating the response                                                           
Everytime the LLM predicts the next word, it has to process everything that came before it. If a model has a context window of 8000 tokens means roughly 6000 words, it will completely forget the very first word of your prompt once the conversation exceeds that limit.

## Transformer Architecture

Before Transformers, models read text sequentially, word by word like recurrent neural net work. This was slow and caused models to forget the beginning of a sentence by the time they reached the end.                        
Transformers process all tokens in a sequence simultaneously, they consist of encoder which reads and understands input and a decoder which generates the output

## Attention mechanism (self-Attention)
This is the "secret sauce" inside the transformer architecture.
It allows the model to weigh the importance of different words in a sequence relative to each other establishing context.                                       
When the model looks at a specific word, the attention mechanism calculates how strongly every other words in the sentence should influence the current word. It creates a web of connections, understanding the adjectives modify specific nouns, and pronouns refer back to specific subject


## LLM Fundamentals
An LLM does not think or reasom like human, it is a massive probability engine.
By training on petabytes of human text, it has mathematically mapped out the relationships between billons of concepts. When we as it a question, it is calculating the statistically most likely sequence of tokens that should follow your input
